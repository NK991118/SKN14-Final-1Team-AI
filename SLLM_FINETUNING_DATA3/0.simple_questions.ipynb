```python
import os
import json
import re
from dotenv import load_dotenv
from openai import OpenAI

# 환경 설정
load_dotenv()

# OpenAI 클라이언트 초기화
client = OpenAI()

def generate_smalltalk_questions():
    prompt = (
        "사내 직원이 챗봇에게 던질 수 있는 캐주얼 인사/잡담/가벼운 추천형 문장을 한국어로 정확히 25개 생성해 주세요. "
        "사람에게 할 만한 질문은 만들지 말고, 무생물(컴퓨터)에게 할 만한 질문으로 생성하세요."
        "문서·보고서·회사자료 요청이나 기능 설명은 절대 포함하지 마세요. "
        "사람의 개인 경험을 묻는 표현(예: '너 점심 뭐 먹었어?')은 제외하고, 대신 추천·감상·잡담 위주로 작성하세요. "
        "각 항목은 1문장, 자연스러운 구어체. "
        "각 질문은 '질문 1: [질문 내용]', '질문 2: [질문 내용]' 형식으로 작성해 주세요. "
    )

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=1500,
        temperature=0.9,
    )
    return response.choices[0].message.content.strip()
```

```python
response_list = []

for i in range(6):
    response = generate_smalltalk_questions()
    question_list = re.findall(r'질문 \d+: (.*)', response)
    response_list += question_list
```

```python
def append_question_list_to_jsonl(question_list, output_jsonl: str):
    with open(output_jsonl, 'a', encoding='utf-8') as f:
        for q in question_list:
            q = (q or "").strip()
            if q:
                json.dump({"question": q}, f, ensure_ascii=False)
                f.write('\n')
```

```python

output_jsonl_list = ["company_questions_cto.jsonl", "company_questions_backend.jsonl", "company_questions_data_ai.jsonl", "company_questions_frontend.jsonl"]   # 결과를 저장할 JSONL 파일 경로
# 예시 실행

for output_jsonl in output_jsonl_list:
    append_question_list_to_jsonl(response_list, output_jsonl)
```
